```python
# Use selenium webdriver to extract content
def render(url):
    driver = webdriver.Chrome(executable_path = './chromedriver')
    driver.get(url)
    time.sleep(3)
    r = driver.page_source
    driver.quit()
    return r
```

```python
# Generate lists for items and prices
items = []
prices = []
for i in range(len(urls)):
    page = render(urls[i])
    soup = BeautifulSoup(page, 'html.parser')
    for item in soup.find_all('div', attrs = {'class': 'a-section a-spacing-none a-spacing-top-small s-title-instructions-style'}):
        item = item.h2.a.text.strip()
        items.append(item)
    for price in soup.find_all(lambda tag: tag.name == 'span' and tag.get('class') == ['a-color-base']):
        price = price.text.strip()
        if '$' in price:
            prices.append(price)
```

```python
df1 = pd.DataFrame()
df1['Item'] = items
df1
```

```python
df2 = pd.DataFrame()
df2['Price'] = prices
df2
```

```python
# Export a csv file containing the features
# NOTE: the features were copied and pasted to the Bik data
compression_opts = dict(method = 'zip', archive_name = 'items.csv')  
df1.to_csv('items.zip', index = False, compression = compression_opts)
df2.to_csv('prices.zip', index = False, compression = compression_opts)  
```
